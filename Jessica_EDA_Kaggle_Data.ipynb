### Fresh Groundwater Abstracted

```python
# Replace 'your_file.csv' with the actual path to your CSV file
file_path = '/Users/ziada/neuefische/capstone-project-AEJS/kaggle/Fresh groundwater abstracted.csv'

# Read the CSV file into a DataFrame
groundwater = pd.read_csv(file_path)

# Display the DataFrame#
print(groundwater)
```

```python
groundwater
```

```python
#Define how many NaN values are in the table?

nan_count_per_column = groundwater.isna().sum()
nan_count_per_column
```

```python
# Replace '...' with the value you want to count
value_to_count = '...'

# Count occurrences of the specified value in the entire DataFrame
count_of_value = (groundwater == value_to_count).sum().sum()

# Display the count
print(f"The value '{value_to_count}' occurs {count_of_value} times in the DataFrame.")
```

```python
print("Groundwater info:")
print(groundwater.info())
```

```python
print("Descriptive statistics:")
groundwater.describe()
```

```python
# Dropping 'Footnotes' column
columns_to_drop2 = [
    'Footnotes 1990', 'Footnotes 1991', 'Footnotes 1992', 'Footnotes 1993',
    'Footnotes 1994', 'Footnotes 1995', 'Footnotes 1996', 'Footnotes 1997',
    'Footnotes 1998', 'Footnotes 1999', 'Footnotes 2000', 'Footnotes 2001',
    'Footnotes 2002', 'Footnotes 2003', 'Footnotes 2004', 'Footnotes 2005',
    'Footnotes 2006', 'Footnotes 2007', 'Footnotes 2008', 'Footnotes 2009',
    'Footnotes 2010', 'Footnotes 2011', 'Footnotes 2012', 'Footnotes 2013',
    'Footnotes 2014', 'Footnotes 2015', 'Footnotes 2016', 'Footnotes 2017'
]

# Drop the specified columns
groundwater = groundwater.drop(columns=columns_to_drop2)
```

```python
print("Column labels:")
print(groundwater.columns)
```

```python
# Drop columns with NaN values
groundwater_cleaned = groundwater.dropna(axis=1)

# Display the DataFrame after dropping columns with NaN values
print(groundwater_cleaned)
```
